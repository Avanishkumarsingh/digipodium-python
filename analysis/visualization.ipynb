```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
iris = px.data.iris()
```

```python
iris.columns.tolist()
```

```python
iris.sepal_length
```

```python
iris.sepal_length.plot()
```

```python
themes = plt.style.available
print(themes)
```

```python
for theme in themes:
    plt.style.use(theme)
    iris.sepal_length.plot(title=f'{theme} theme')
    plt.show()
```

```python
# theme set
plt.style.use('seaborn-v0_8-ticks')
```

```python
iris.sepal_length.plot(figsize=(20,5))
```

```python
iris.sepal_length.plot(figsize=(15,5), 
    color= 'red',
    marker = 'o',
    markersize =8,
    alpha=.5)
```

```python
iris[['sepal_length','sepal_width']]. plot(figsize=(15,5), subplots=True)
```

```python
iris[['sepal_length', 'sepal_width']].plot(figsize=(15,5), subplots=True, layout=(1,2))
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris[cols].plot(figsize=(10,10), subplots=True, layout=(2,2))
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

```python
iris.plot(kind='scatter',
     x='sepal_length',
     y='sepal_width',
     c='species_id',
     colormap='rainbow')
```

```python
iris.plot(kind='scatter',
     x='petal_length',
     y='petal_width',
     c='species_id',
     colormap='rainbow',
     s= 60)
```

```python
iris.plot(kind='hexbin',
     x= 'sepal_length',
     y= 'sepal_width',
     gridsize= 15,
     figsize= (10,5))
```

```python
iris.plot(kind='scatter',
     x= 'petal_length',
     y= 'sepal_width',
     c= 'species_id',
     colormap='rainbow',
     s= 100)
```

```python
fig = px.line(iris, y='sepal_width')
fig.show()
```

```python
fig = px.scatter(iris, x='petal_length', y='petal_width', color='species')
fig.show()
```

```python
tips = px.data.tips()
tips
```

```python
tips.sex.value_counts()
```

```python
tips.sex.value_counts().plot(kind='pie',
                            autopct= '%.1f%%',
                            colors=['black','red'],
                            textprops={'color': 'w',
                                         'fontsize':20},
                            wedgeprops={'width':0.5},
                            pctdistance=0.7,
                            )
```

```python
tips.day.value_counts().plot(kind='bar', figsize=(6,4), title='Day wise comparison')
plt.savefig('daywise.png', dpi=300, bbox_inches='tight', pad_inches=2)
```

```python
tips.day.value_counts().plot(
    kind='bar',
    figsize=(6,4),
    hatch='//',
    edgecolor='white',
    title='Day Wise Comparison'
)
      
```

```python
tips.total_bill.plot(kind='hist',
                figsize=(6,4),
                title='Total Bill Distribution',
                bins=20,
                color='red',
                edgecolor='black',
                alpha=.5)
```

```python
tips.total_bill.plot.box(
    figsize=(6,4),
    showmeans=True,
    meanline=True,
    meanprops={'color':'green', 'linewidth':.5, 'linestyle':'--'},
    notch=True
)
plt.show()
```

```python
px.box(tips, x='day', y='total_bill')
```
